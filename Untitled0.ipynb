### Preprocesing

```python
# remove the numbers inside paramtheisis

def paranthesis(text):
  text =  re.sub(r"\([0-9]+\)", "", text)
  return text
df['itemname'] = df['itemname'].apply(paranthesis)

df
```

```python
# Stop words removal

stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
df['itemname'] = df['itemname'].apply(remove_stopwords)

df
```

```python
#removing unicode charactors
def remove_unicode_chars(text):
    text = text.encode("ascii", "ignore").decode()
    return text
df['itemname'] = df['itemname'].apply(remove_unicode_chars)

df
```

```python
#removing punctuations

# string.punctuation
def remove_punctuations(text):
    text = re.sub('[%s]' % re.escape(string.punctuation), " ",text)
    return text
df['itemname'] = df['itemname'].apply(remove_punctuations)
df
```

```python
#converting emojis
def convert_emojis(text):
    for emot in UNICODE_EMOJI:
        text = text.replace(emot, "_".join(UNICODE_EMOJI[emot].replace(",","").replace(":","").split()))
    return text
df['itemname'] = df['itemname'].apply(convert_emojis)
df
```

```python
#removing extra spaces
def remove_extra_spaces(text):
    text = re.sub(' +', ' ', text).strip()
    return text
df['itemname'] = df['itemname'].apply(remove_extra_spaces)

df
```

```python
# converting capital letters to simple letters

df['itemname'] = df['itemname'].str.lower()
df
```

```python
#lemmatization
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    words = [lemmatizer.lemmatize(word) for word in text.split()]
    text = ' '.join(words)
    return text
df['itemname'] = df['itemname'].apply(lemmatize_text)

df
```

### Feature Engineering

```python
# tokenizing the data
token = Tokenizer()
token.fit_on_texts(df['itemname'])
```

```python
seq = token.texts_to_sequences(df['itemname'])
```

```python
# pad the sentences

max_length = 50
pad_seq = pad_sequences(seq, maxlen=max_length, padding='post')
```

```python
embedding_vector = {}
vocab_size = len(token.word_index) +1

f = open('/content/drive/MyDrive/Pickme Assignment/glove.6B.50d.txt')

for line in tqdm(f):
  value = line.split(' ')
  word = value[0]
  coef = np.array(value[1:], dtype = 'float32')
  embedding_vector[word] = coef
```

```python
#create a weight matrix for words in data

embedding_matrix = np.zeros((vocab_size, 50))
for word,i in tqdm(token.word_index.items()):
    embedding_value = embedding_vector.get(word)
    if embedding_value is not None:
        embedding_matrix[i] = embedding_value
```

```python
#encode the label
label_encoder = preprocessing.LabelEncoder()
df['tag']= label_encoder.fit_transform(df['tag'])
y = to_categorical(df['tag'])
```

```python
df['tag'].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split( pad_seq, y, test_size=0.33, random_state=100)
```

### Modeling

```python

model = Sequential()
model.add(Embedding(vocab_size,50, weights = [embedding_matrix], input_length=max_length, trainable=True))
model.add(Bidirectional(LSTM(25, dropout=0.1)))
model.add(Dense(25, activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.1))
model.add(Dense(12, activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.1))
model.add(Dense(38,activation = 'softmax'))
model.summary()

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.3,
                                                 patience=2, verbose=1, min_lr=1e-7)
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

model.compile(
    optimizer = optimizer,
    loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True),
    #metrics = [tf.keras.metrics.Accuracy(), tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
    metrics = [tf.keras.metrics.BinaryAccuracy(name='accuracy')]
)



history = model.fit(
    X_train, y_train,
    batch_size = 32,
    epochs = 10,
    validation_split = 0.3,

    callbacks = [reduce_lr]
)
```

```python
# get the predictions
y_pred = model.predict(X_test)
```

```python
# get the position for highest probability
y_pred = np.argmax(y_pred, axis=1)
```
